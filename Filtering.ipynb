```python
import pandas as pd

path = "US_Accidents_March23.csv"
df = pd.read_csv(path)

# pd.set_option('display.max_rows', None)
# df
# df.head()
```

<!-- Source, End_Lat, End_Lng, Distance(mi), Country, Weather_Timestamp, Civil_Twilight, Astronomical_Twilight -->

```python
df_filtered = df.copy()
del df_filtered["Source"]
del df_filtered["End_Lat"]
del df_filtered["End_Lng"]
del df_filtered["Distance(mi)"]
del df_filtered["Country"]
del df_filtered["Weather_Timestamp"]
del df_filtered["Civil_Twilight"]
del df_filtered["Astronomical_Twilight"]
del df_filtered["Nautical_Twilight"]
# df_filtered
```

```python
df_filtered['Start_Time'] = pd.to_datetime(df_filtered['Start_Time'], format ='mixed')
df_filtered['End_Time'] = pd.to_datetime(df_filtered['End_Time'], format ='mixed')
```

```python
df_filtered.loc[:, 'Year'] = df_filtered['Start_Time'].dt.year
df_filtered.loc[:, 'Month'] = df_filtered['Start_Time'].dt.month
df_filtered.loc[:, 'Day_Name'] = df_filtered['Start_Time'].dt.day_name()
```

```python
# df_filtered.head()
```

```python
df_2016 = df_filtered[df_filtered["Year"] == 2016]
df_2017 = df_filtered[df_filtered["Year"] == 2017]
df_2018 = df_filtered[df_filtered["Year"] == 2018]
df_2019 = df_filtered[df_filtered["Year"] == 2019]
df_2020 = df_filtered[df_filtered["Year"] == 2020]
df_2021 = df_filtered[df_filtered["Year"] == 2021]
df_2022 = df_filtered[df_filtered["Year"] == 2022]
df_2023 = df_filtered[df_filtered["Year"] == 2023]
```

```python
df_2016.to_csv('Data/df_2016.csv', index=False)
df_2017.to_csv('Data/df_2017.csv', index=False)
df_2018.to_csv('Data/df_2018.csv', index=False)
df_2019.to_csv('Data/df_2019.csv', index=False)
df_2020.to_csv('Data/df_2020.csv', index=False)
df_2021.to_csv('Data/df_2021.csv', index=False)
df_2022.to_csv('Data/df_2022.csv', index=False)
df_2023.to_csv('Data/df_2023.csv', index=False)
```
