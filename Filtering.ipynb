# WEATHER AND ENVIROMENT

```python
windy_conditions = df_filtered[df_filtered['Weather_Condition'].str.contains('Windy', na=False)]
rainy_conditions = df_filtered[df_filtered['Weather_Condition'].str.contains('Rain', na=False)]
foggy_conditions = df_filtered[df_filtered['Weather_Condition'].str.contains('Fog', na=False)]
snow_conditions = df_filtered[df_filtered['Weather_Condition'].str.contains('Snow', na=False)]
```

```python
rainy_perc = (rainy_conditions["ID"].count() / df_filtered["ID"].count())* 100
windy_perc = (windy_conditions["ID"].count() / df_filtered["ID"].count())* 100
foggy_perc = (foggy_conditions["ID"].count() / df_filtered["ID"].count())* 100
snow_perc = (snow_conditions["ID"].count() / df_filtered["ID"].count())* 100
```

```python
print(f"Occurance of accidents when its Raining : {rainy_perc:.2f}%")
print(f"Occurance of accidents when its Windy : {windy_perc:.2f}%")
print(f"Occurance of accidents when its Foggy : {foggy_perc:.2f}%")
print(f"Occurance of accidents when its Snowing : {snow_perc:.2f}%")
```

```python
print(f"Average severity of accidents when its Windy : {windy_conditions["Severity"].mean():.2f}")
print(f"Average severity of accidents when its Rainy : {rainy_conditions["Severity"].mean():.2f}")
print(f"Average severity of accidents when its Foggy : {foggy_conditions["Severity"].mean():.2f}")
print(f"Average severity of accidents when its Snowing : {snow_conditions["Severity"].mean():.2f}")
```

# TIME ANALYSIS

```python
total = df_filtered.loc[(df_filtered["Sunrise_Sunset"] == "Day") | (df_filtered["Sunrise_Sunset"] == "Night")]["ID"].count()
day_df = df_filtered[df_filtered["Sunrise_Sunset"] == "Day"]
day_acc = day_df["ID"].count()
night_df = df_filtered[df_filtered["Sunrise_Sunset"] == "Night"]
night_acc = night_df["ID"].count()
```

```python
day_per = (day_acc/total)*100
night_per = (night_acc/total)*100
print(f"Percentage of the accidents happened during day : {day_per:.2f}%")
print(f"Percentage of the accidents happened during night : {night_per:.2f}%")
```

```python
# Data
# day_per = 69.23
# night_per = 30.77
labels = ['Day', 'Night']
percentages = [day_per, night_per]
colors = ['gold', 'midnightblue']
# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    percentages,
    labels=labels,
    colors=colors,
    autopct='%1.2f%%',
    startangle=140,
    textprops={'color': 'white'}
)
# Formatting
plt.title('Accidents Distribution: Day vs Night', color='black', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.show()
```

```python
day_df = df_filtered.copy()
day_counts = day_df.groupby('Day_Name').size().reset_index(name='Count')

# Sort the results by day of the week
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_counts['Day_Name'] = pd.Categorical(day_counts['Day_Name'], categories=day_order, ordered=True)
day_counts = day_counts.sort_values('Day_Name')

print(day_counts)
```

```python
#Plot days of week
day_counts.plot(x='Day_Name', y='Count', kind='bar', figsize=(10, 6), color='skyblue')
# Add titles and labels
plt.title('Accident Counts by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
